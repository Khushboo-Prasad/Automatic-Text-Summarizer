```python
def summarize(all_para_data):
    f_data=re.sub(r'\[[0-9]*\]',' ',all_para_data)          # removing all square brackets
    data=re.sub(r'\s+',' ',f_data)                          # concatenating all the paragraphs
    #print(data)

    sent_tokens=nltk.sent_tokenize(data)
    word_tokens=nltk.word_tokenize(data)
    #print(word_tokens)

    
    from nltk.corpus import stopwords
    stopwords=nltk.corpus.stopwords.words('english')   # Stopwords are the English words which does not add much meaning to a sentence. 
                                                       #They can safely be ignored without sacrificing the meaning of the sentence. 
    word_frequencies={}
    for word in word_tokens:
        if word not in stopwords:
            if word not in word_frequencies.keys():    # keys are the words and value is their count
                word_frequencies[word]=1
            else:
                word_frequencies[word]+=1              # not in stopwords but in keys
    #print(word_frequencies)

    max_freq_word=max(word_frequencies.values())
    #print(max_freq_word)
    for word in word_frequencies.keys():
        word_frequencies[word]=(word_frequencies[word]/max_freq_word)
    #print(word_frequencies)
    sentence_score={}
    for sentence in sent_tokens:                              # .lower() converts all the uppercase to lowercase
        for word in nltk.word_tokenize(sentence.lower()):     # extracting every word from the passed sentence token  
            if word in word_frequencies.keys():
                if(len(sentence.split(' '))<30):              
                    if sentence not in sentence_score.keys():
                        sentence_score[sentence] = word_frequencies[word]
                    else:
                        sentence_score[sentence] += word_frequencies[word]
    #print(sentence_score)
    import heapq
    summary=heapq.nlargest(10,sentence_score,key=sentence_score.get)
    show_text(summary)
    
    
def show_text(summary):
         for i in summary:
            print(i,end=' ')
            #clear_text2()
            t12.insert(tk.END,i) or t42.insert(tk.END,i) or t22.insert(tk.END,i) or t32.insert(tk.END,i)

        
def geturl():
    global url
    url=e41.get()
    #print(url)
    htmldoc=requests.get(url).content
    soup_object=bs(htmldoc,'html.parser')
    para_content=soup_object.find_all('p')
    all_para_data=""
    for content in para_content:
         all_para_data+=content.text
    summarize(all_para_data)
    
def get_text_data1():
    data=t11.get("1.0",END)
    if data:
        summarize(data)
       
    else:
         messagebox.showinfo("No text found!!") 
    
def get_text_data2():
    data=t21.get("1.0",END)
    summarize(data)

def get_image_data3():
    summarize(img_text)
    
     
def clear_text2():
    t11.delete('1.0',tk.END) or e41.delete(0,tk.END) or t21.delete('1.0',tk.END) or t31.delete('1.0',tk.END)
        
def clear_results():
    t12.delete('1.0',tk.END) or t22.delete('1.0',tk.END) or t32.delete('1.0',tk.END) or t42.delete('1.0',tk.END)

def upload():
    file = askopenfile(mode='r')
    data = file.read()
    t21.insert(tk.END,data)
    print(data)
    
    
def text_extraction(imgloc):
    
    img=cv2.imread(imgloc)
    #print(img)
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)                    
    file="{}.png".format(os.getpid())                                            
    cv2.imwrite(file,img) 
    global img_text
    img_text=pytesseract.image_to_string(Image.open(file))                                                                                                      
    os.remove(file)                                                                      
    print(img_text)
    t31.insert(tk.END,img_text)
     
    
def upload_image():
    file = fd.askopenfilename()
    print(file)
    text_extraction(file)
    
def save1():
    data=t12.get("1.0",END)
    #print(data)
    
    f = fd.asksaveasfile(mode='w',defaultextension=".txt",filetypes=[('All Files', '*.*'),('Text Document', '*.txt'),("Word files",".doc")])
    if f is None :
        messagebox.showinfo("Error", "Cancelled!!")   
    else:
        f.write(data)
        f.close()
        messagebox.showinfo("File saved at location :",f)
        
def save2():
    data=t22.get("1.0",END)
    #print(data)
    
    f = fd.asksaveasfile(mode='w',defaultextension=".txt",filetypes=[('All Files', '*.*'),('Text Document', '*.txt'),("Word files",".doc")])
    if f is None :
        messagebox.showinfo("Error", "Cancelled!!")   
    else:
        f.write(data)
        f.close()
        messagebox.showinfo("File saved at location :",f)
    
def save3():
    data=t32.get("1.0",END)
    #print(data)
    
    f = fd.asksaveasfile(mode='w',defaultextension=".txt",filetypes=[('All Files', '*.*'),('Text Document', '*.txt'),("Word files",".doc")])
    if f is None :
        messagebox.showinfo("Error", "Cancelled!!")   
    else:
        f.write(data)
        f.close()
        messagebox.showinfo("File saved at location :",f)
        
def save4():
    data=t42.get("1.0",END)
    #print(data)
    
    f = fd.asksaveasfile(mode='w',defaultextension=".txt",filetypes=[('All Files', '*.*'),('Text Document', '*.txt'),("Word files",".doc")])
    if f is None :
        messagebox.showinfo("Error", "Cancelled!!")   
    else:
        f.write(data)
        f.close()
        messagebox.showinfo("File saved at location :",f)
        
        
def audio1():
    
    data=t12.get("1.0",END)
    language='en'
    my = gTTS(text=data,lang=language,slow=False)
    date_string = dt.datetime.now().strftime("%d%m%Y%H%M%S")
    filename = "audio"+date_string+".mp3"
    my.save(filename)

    mixer.init()
    mixer.music.load(filename)
    mixer.music.play()
        
def audio2():
    data=t22.get("1.0",END)
    language='en'
    my = gTTS(text=data,lang=language,slow=False)
    date_string = dt.datetime.now().strftime("%d%m%Y%H%M%S")
    filename = "audio"+date_string+".mp3"
    my.save(filename)

    mixer.init()
    mixer.music.load(filename)
    mixer.music.play()
        
def audio3():
    data=t32.get("1.0",END)
    language='en'
    my = gTTS(text=data,lang=language,slow=False)
    date_string = dt.datetime.now().strftime("%d%m%Y%H%M%S")
    filename = "audio"+date_string+".mp3"
    my.save(filename)

    mixer.init()
    mixer.music.load(filename)
    mixer.music.play() 

def audio4():
    data=t42.get("1.0",END)
    language='en'
    my = gTTS(text=data,lang=language,slow=False)
    date_string = dt.datetime.now().strftime("%d%m%Y%H%M%S")
    filename = "audio"+date_string+".mp3"
    my.save(filename)

    mixer.init()
    mixer.music.load(filename)
    mixer.music.play()
        
#####################################################################################
# main Window
root=tk.Tk()
root.title("Your Own Summarizer")
root.geometry("800x800")

# notebook
note=ttk.Notebook(root,width=800,height=800)
note.pack()

###################################################################################
# making Frames and adding them into the notebook
f1=Frame(note,bg='navajowhite2')
f1.pack(expand=0)
note.add(f1,text="Home")

# adding text widget to frame 1
t11=Text(f1,height=20,width=100,bg='bisque')
t11.grid(row=0,column=0,columnspan=5,padx=5,pady=20)

t12=Text(f1,height=20,width=100,bg='bisque')
t12.grid(row=2,column=0,columnspan=5,padx=5,pady=20)


# adding buttons to the frame 1
b11=Button(f1,text='Upload',bg='gold',width=10)
b11.grid(row=1,column=0)

b12=Button(f1,text='Summarise',bg='magenta2',width=10,command=get_text_data1)
b12.grid(row=1,column=1)

b13=Button(f1,text='Reset',bg='lawngreen',width=10,command=clear_text2)
b13.grid(row=1,column=2)

b14=Button(f1,text='Save',bg='cyan2',width=10,command=save1)
b14.grid(row=1,column=3)

b15=Button(f1,text='Clear results',bg='darkorange',width=10,command=clear_results)
b15.grid(row=1,column=4)

b16=Button(f1,text='Listen', bg='yellow',width=10,command=audio1)
b16.grid(row=1,column=0)

l11=Label(f1,text="Enter your text below : ",bg='bisque1')
l11.grid(row=0,column=0,sticky=(N,W))

######################################################################################

f2=Frame(note,bg='lightgoldenrod1')
f2.pack(expand=1)
note.add(f2,text="File")

# adding text widget to frame 2
t21=Text(f2,height=20,width=100,bg='bisque')
t21.grid(row=0,column=0,columnspan=7,padx=5,pady=20)

t22=Text(f2,height=20,width=100,bg='bisque')
t22.grid(row=2,column=0,columnspan=7,padx=5,pady=20)


# adding buttons to the frame 2
b21=Button(f2,text='Upload',bg='gold',width=10,command=upload)
b21.grid(row=1,column=0)

b22=Button(f2,text='Summarise',bg='magenta2',width=10,command=get_text_data2)
b22.grid(row=1,column=1)

b23=Button(f2,text='Reset',bg='lawngreen',width=10,command=clear_text2)
b23.grid(row=1,column=2)

b24=Button(f2,text='Save',bg='cyan2',width=10,command=save2)
b24.grid(row=1,column=5)

b25=Button(f2,text='Clear results',bg='darkorange',width=10,command=clear_results)
b25.grid(row=1,column=4)

b26=Button(f2,text='Listen', bg='yellow',width=10,command=audio2)
b26.grid(row=1,column=3)

l11=Label(f2,text="Enter your text below : ",bg='bisque')
l11.grid(row=0,column=0,sticky=(N,W))
################################################################################################

f3=Frame(note,bg='pink')
f3.pack(expand=1)
note.add(f3,text="Image")

# adding text widget to frame 3
t31=Text(f3,height=20,width=100,bg='bisque')
t31.grid(row=0,column=0,columnspan=6,padx=5,pady=20)

t32=Text(f3,height=20,width=100,bg='bisque')
t32.grid(row=2,column=0,columnspan=6,padx=5,pady=20)


# adding buttons to the frame 3
b31=Button(f3,text='Upload',bg='gold',width=10,command=upload_image)
b31.grid(row=1,column=0)

b32=Button(f3,text='Summarise',bg='magenta2',width=10,command=get_image_data3)
b32.grid(row=1,column=1)

b33=Button(f3,text='Reset',bg='lawngreen',width=10,command=clear_text2)
b33.grid(row=1,column=5)

b34=Button(f3,text='Save',bg='cyan2',width=10,command=save3)
b34.grid(row=1,column=3)

b35=Button(f3,text='Clear results',bg='darkorange',width=10,command=clear_results)
b35.grid(row=1,column=4)

b36=Button(f3,text='Listen', bg='yellow',width=10,command=audio3)
b36.grid(row=1,column=2)

l11=Label(f3,text="Enter your text below : ",bg='bisque')
l11.grid(row=0,column=0,sticky=(N,W))
##############################################################################################

f4=Frame(note,bg='blue4')
f4.pack(expand=1)
note.add(f4,text="Url")


# adding  widgets to frame 4

l41=Label(f4,text="Enter the URL")
l41.grid(row=1,column=0,pady=50)

e41=Entry(f4,width=80,bg='gray80')
e41.grid(row=1,column=1,pady=50,columnspan=3 )

t42=Text(f4,height=20,width=100,bg='bisque')
t42.grid(row=5,column=0,columnspan=5,padx=5,pady=20)


# adding buttons to the frame 4
b42=Button(f4,text='Summarise',bg='magenta2',width=10,command=geturl)
b42.grid(row=4,column=1)

b43=Button(f4,text='Reset',bg='lawngreen',width=10,command=clear_text2)
b43.grid(row=4,column=3)

b44=Button(f4,text='Save',bg='cyan2',width=10,command=save4)
b44.grid(row=4,column=2)

b45=Button(f4,text='Clear results',bg='darkorange',width=10,command=clear_results)
b45.grid(row=4,column=4)

b46=Button(f4,text='Listen', bg='yellow',width=10,command=audio4)
b46.grid(row=4,column=0)



root.mainloop()
```
